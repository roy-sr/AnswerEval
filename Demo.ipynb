## Prepare the reference answer(s) intermediate files

### If preparing for new Question then it might take some time.

```python
df_sheet_reference = pd.read_excel('dataFile.xlsx', sheet_name='reference', usecols = "A:C",index_col=0)
```

```python
for index, row in df_sheet_reference.iterrows():

    app.prepare_reference_ans(index, row['Answer'], row['Question'])

print('Ref ans processing complete')
```

### evaluate the reference answer and get haystack co-efficient

```python
haystack_coeff_dict = {}

for index, row in df_sheet_reference.iterrows():
    
    print(index)
    print(row['Answer'])
    haystack_coeff, _ =  app.evaluate_candidate_ans(refAnsID=index, candAnsID=index, \
                                                             candAnstext=row['Answer'], haystack_coeff = None)
     
    haystack_coeff_dict[index]  =haystack_coeff
    
    
```

```python
df_sheet_canditateAns = pd.read_excel('dataFile.xlsx', sheet_name='canditateAns', usecols = "A:C",index_col=0)
```

```python
use = USE()
app = App(use)
finale_score_list = []

for index, row in df_sheet_canditateAns.iterrows():

    _ , score = app.evaluate_candidate_ans(refAnsID=row['RefQID'], candAnsID=index, \
                                   candAnstext=row['Answer'], haystack_coeff = haystack_coeff_dict[row['RefQID']])
    
    finale_score_list.append(index + ' ' + str(score))
```

### Display the scores

```python
for temp_item in finale_score_list:
    print(temp_item)
```
